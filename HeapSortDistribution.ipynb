# Experimental analysis of the Heap Sort execution time 

```
def max_heap(a,index):
  izq, der= 2*index,(2*index)+1
  if izq < len(a) and a[izq] > a[index]:
    max = izq
  else:
    max = index
  if der < len(a) and a[der] > a[max]:
    max = der
  if max != index:
    a[index], a[max] = a[max],a[index]
    max_heap(a,max)
def build_max_heap(a):
  for i in range(int(len(a) / 2), -1, -1):
    max_heap(a, i)
def heap_sort(a):
  sorted = []
  build_max_heap(a)
  while len(a) > 1:
    maxim = a[1]
    a[1] = a[-1]
    del a[-1]
    sorted.insert(0,maxim)
    max_heap(a,1)
  if len(a) == 1:
    sorted.append(a[0])
  return sorted
```

# Insertion sort  time complexity histograms for permutations of $n$ different elements counting steps like CLRS I.2.1 

$$t (n) =  2n - 1  + 3 \sum_{j=2,..,n} t_j $$

## Best case - increasingly sorted array 
$$t_b(n) =  5n - 4 = \Theta(n)$$

## Worst case - de creasingly sorted array 

$$t_w(n) = \frac{3}{2} n^2 + \frac{7}{2} n  - 4 = \Theta(n^2)$$

## Average case - expected number of comparisons + swaps on random unifrom permutation of $n$ different elements 

$$t_a(n) = \frac{3}{4} n^2 + \frac{17}{4} n  - 4  = \Theta(n^2)$$

 

```
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1,21,1)

plt.plot(n, 5*n-4,  label = 'Heap sort best time $5n-4$')
plt.plot(n, 3/4*n**2+17/4*n-4, label = 'Heap average time $3/4 n*^2+17/4 n - 4$')
plt.plot(n, 3/2*n**2+7/2*n-4, label = 'Heap average time $3/2 n*^2+7/2 n - 4$')
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc = 'upper left')
```

```
def max_heap_steps(a,index,steps):
  izq, der= 2*index,(2*index)+1
  steps = 0
  if izq < len(a) and a[izq] > a[index]:
    max = izq
    steps += 1
  else:
    max = index
    steps += 1
  if der < len(a) and a[der] > a[max]:
    max = der
    steps += 1
  if max != index:
    a[index], a[max] = a[max],a[index]
    steps += 2
    steps += max_heap_steps(a,max,steps)
  return steps


def build_max_heap_steps(a,steps):
  for i in range(int(len(a) / 2), -1, -1):
    steps += max_heap_steps(a, i,steps)
  return steps

def heap_sort_steps(a):
  aux = len(a)
  steps = 1
  sorted = []
  steps += build_max_heap_steps(a,steps)
  while aux > 1:
    maxim = a[1]
    a[1] = a[-1]
    #del a[-1]
    aux-=1
    sorted.insert(0,maxim)
    steps += max_heap_steps(a,1,steps)
    steps += 6
  if len(a) == 1:
    sorted.append(a[0])
    steps +=2
  return steps
```
